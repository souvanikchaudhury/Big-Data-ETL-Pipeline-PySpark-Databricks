```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

#Trimming 

```python
df = spark.table("workspace.bronze.crm_cust_info")
```

```python
for field in df.schema.fields:
    if isinstance(field.dataType,StringType):
        df = df.withColumn(field.name,trim(field.name))
```

```python
df.display()
```

#Normalisation

```python
df = df.withColumn(
    "cst_marital_status",
    when(col("cst_marital_status").isNull(), "Unknown")
    .when(upper(trim(col("cst_marital_status"))) == "M", "Married")
    .when(upper(trim(col("cst_marital_status"))) == "S", "Single")
    .otherwise("Unknown")
)
```

```python
df.display()
```

```python
df = df.withColumn(
    "cst_gndr",
    when(col("cst_gndr").isNull(), "Unknown")
    .when(upper(trim(col("cst_gndr"))) == "M", "Male")
    .when(upper(trim(col("cst_gndr"))) == "F", "Female")
    .otherwise("Unknown")
)
```

```python
df.display()
```

```python
df = df.filter(col("cst_id").isNotNull())
```

```python
RENAME_MAP = {
    "cst_id": "customer_id",
    "cst_key": "customer_number",
    "cst_firstname": "first_name",
    "cst_lastname": "last_name",
    "cst_marital_status": "marital_status",
    "cst_gndr": "gender",
    "cst_create_date": "created_date"
}

for old_name, new_name in RENAME_MAP.items():
    df = df.withColumnRenamed(old_name, new_name)
```
